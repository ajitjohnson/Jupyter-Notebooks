```python
# try OTSU thresholding
for i in range(5):
    #i = 84
    #img  = ((pos[0, i, ...]/65535 )*255).astype('uint8')
    #img  = (pos[0, i, ...]/256).astype('uint8')
    img = ((pos[0, i, ...]/np.percentile(pos[0, i, ...], 99.99) )*255).astype('uint8')

    ret,th1 = cv.threshold(img,2,24,cv.THRESH_BINARY)
    th2 = cv.adaptiveThreshold(img,24,cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY,5,0)
    ret3,th3 = cv.threshold(img,0,255,cv.THRESH_OTSU)
    titles = ['Original Image', 'Global Thresholding (v = 127)', 'Adaptive Mean Thresholding', 'OTSU Thresholding']
    images = [img, th1, th2, th3]
    for i in range(4):
        plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
        plt.title(titles[i])
        plt.xticks([]),plt.yticks([])
    plt.show()
    clear_output(wait = True)
    plt.pause(2)
```

```python
# identify the pos and neg cells for marker of interest
def pos_neg_cells (adata, marker, threshold=[2, 12, 88, 98], label='pos_neg'):
    data = pd.DataFrame(adata.X, columns=adata.var.index, index=adata.obs.index)
    # preprocess data
    def pre_process (data):
        # clip outliers
        def clipping (x):
            clip = x.clip(lower =np.percentile(x,0.01), upper=np.percentile(x,99.99)).tolist()
            return clip
        processsed_data = data.apply(clipping)
        processsed_data = np.log1p(processsed_data) # log transform the data
        return processsed_data
    processsed_data = pre_process (data)
    # process the clipped data
    marker_of_interest = processsed_data[marker]
    # determine the threshold values
    threshold.sort()
    low_a = np.percentile(marker_of_interest, threshold[0]); low_b = np.percentile(marker_of_interest, threshold[1])
    high_a = np.percentile(marker_of_interest, threshold[2]); high_b = np.percentile(marker_of_interest, threshold[3])
    # identify the indeces of cells that fall within the threshold
    neg = np.where(marker_of_interest.between(low_a, low_b))
    pos = np.where(marker_of_interest.between(high_a, high_b))
    # replace the values
    marker_of_interest.iloc[neg] = 'neg'
    marker_of_interest.iloc[pos] = 'pos'
    marker_of_interest = marker_of_interest.apply(lambda x: x if type(x) == str else np.nan)
    marker_of_interest = marker_of_interest.fillna('rest')
    # add the results to anndata
    marker_of_interest = marker_of_interest.reindex(adata.obs.index)
    adata.obs[label] = marker_of_interest
    # return 
    return adata
```

```python
marker = 'CD3D'
```

```python
adata = pos_neg_cells (adata, marker=marker, threshold=[5, 10, 90, 95])
adata.obs['pos_neg'].value_counts()
```
